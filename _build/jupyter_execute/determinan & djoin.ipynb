1. A = $$\begin{bmatrix} -7 & -5 \\ 1 & 4 \end{bmatrix}$$

   jawaban :

   Ekspansi baris pertama:   $$det(A)=(тИТ1)1+1(тИТ7)тЛЕM11тАЛ+(тИТ1)1+2(тИТ5)тЛЕM12тАЛ$$

   - $M11 = 4$
   - $M12 = 1$

   $$
   =(+1)(тИТ7)(4)+(тИТ1)(тИТ5)(1)\\
   =тИТ28+5=тИТ23
   $$

2. $$A = \begin{bmatrix} 0 & 2 & -3 \\ 1 & -2 & -1 \\ 0 & 0 & 1 \end{bmatrix}$$

   jawaban:

   Ekspansi baris ke-3 :  

   $$
   \sum_{k=1}^3 (-1)^{3+k} a_{3k} {M_3k}
   $$


   elemen baris ke3 :   0, 0, 1

   Hitung ekspansi :  $$det(A)=(тИТ1)3+1(0)тЛЕM31тАЛ+(тИТ1)3+2(0)тЛЕM32тАЛ+(тИТ1)3+3(1)тЛЕM33тАЛ$$
   $$
   (тИТ1)4(0)(M31тАЛ)=0 \\
   (тИТ1)5(0)(M32тАЛ)=0 \\
   (тИТ1)6(1)(M33)=(+1)(1)тЛЕM33
	тАЛтАЛ$$

   $$
   M_{33} =
   \begin{vmatrix}
   0 & 2 \\
   1 & -2
   \end{vmatrix}
   = (0)(-2) - (2)(1)
   = 0 - 2
   = -2
   $$

   Hasil akhir  :  $$det(A)=0+0+(1)(тИТ2)=тИТ2тАЛ$$
                     $$det(A)=тИТ2$$

3. $$A = \begin{bmatrix} 1 & -3 & 1 & 1 \\ -3 & 1 & 1 & 1 \\ 1 & 1 & -3 & 1 \\ 1 & 1 & 1 & -3 \end{bmatrix}.$$

   jawaban :

   Ekspansi baris ke-4 :
   $$
   \sum_{k=1}^4 (-1)^{4+k} a_{4k} {M_4k}
   $$


   elemen baris ke-4 : 1, 1, 1, -3

   Hitung ekspansi : $$det(A)=(тИТ1)5(1)M41тАЛ+(тИТ1)6(1)M42тАЛ+(тИТ1)7(1)M43тАЛ+(тИТ1)8(тИТ3)M44$$

   Hitung minor : 
   $M_{41}$

   $$
   M_{41} =
   \begin{vmatrix}
   -3 & 1 & 1 \\
   1 & -3 & 1 \\
   1 & 1 & -3
   \end{vmatrix}
   = -16
   $$

   $M_{42}$

   $$
   M_{42} =
   \begin{vmatrix}
   1 & 1 & 1 \\
   -3 & 1 & 1 \\
   1 & -3 & 1
   \end{vmatrix}
   = 16
   $$

   $M_{43}$

   $$
   M_{43} =
   \begin{vmatrix}
   1 & -3 & 1 \\
   -3 & 1 & 1 \\
   1 & 1 & 1
   \end{vmatrix}
   = -16
   $$

   $M_{44}$
   $$
   M_{44} =
   \begin{vmatrix}
   1 & -3 & 1 \\
   -3 & 1 & 1 \\
   1 & 1 & -3
   \end{vmatrix}
   = -16
   $$

   Substitusi :
   $$
   \det(A) =
   (-1)(1)(-16)
   + (1)(1)(16)
   + (-1)(1)(-16)
   + (1)(-3)(-16)
   $$

   Hasil:

   $$
   \det(A) = 16 + 16 + 16 + 48 = 96\\
   \boxed{\det(A) = 96}

B. Matriks Adjoin 

4. $$A = \begin{bmatrix} -7 & -5 \\ 1 & 4 \end{bmatrix}$$

   jawaban :

   Rumus : $$(\operatorname{adj} A)_{ij} = (-1)^{i+j} M_{ji}\\
               A^{-1} = \frac{1}{\det A} \operatorname{adj} A.$$
   
   - Determinan

   $$
   \det(A) = (-7)(4) - (-5)(1)
   = -28 + 5
   = -23
   $$

   - Matriks Kofaktor

   Gunakan tanda:
   $$
   \begin{bmatrix}
   + & - \\
   - & +
   \end{bmatrix}
   $$

    Minor:

   $$
   M_{11} = 4,\quad M_{12} = 1,\quad M_{21} = -5,\quad M_{22} = -7
   $$

    Kofaktor:

   $$
   C =
   \begin{bmatrix}
   (+1)(4) & (-1)(1) \\
   (-1)(-5) & (+1)(-7)
   \end{bmatrix}
   =
   \begin{bmatrix}
   4 & -1 \\
   5 & -7
   \end{bmatrix}
   $$


   - Adjoin (transpose kofaktor)

   $$
   \text{adj}(A) =
   \begin{bmatrix}
   4 & 5 \\
   -1 & -7
   \end{bmatrix}
   $$

   - Invers Matriks

      Gunakan rumus:

   $$
   A^{-1} = \frac{1}{\det(A)} \cdot \text{adj}(A)
   $$

   $$
   A^{-1} = \frac{1}{-23}
   \begin{bmatrix}
   4 & 5 \\
   -1 & -7
   \end{bmatrix}
   $$

  - Hasil Akhir

   $$
   A^{-1} =
   \begin{bmatrix}
   -\frac{4}{23} & -\frac{5}{23} \\
   \frac{1}{23} & \frac{7}{23}
   \end{bmatrix}
   $$
