**1. Sistem Persamaan**

$$
\begin{cases}
x + y + z = 6 \\
2x + y - z = 3 \\
x - y + z = 2
\end{cases}
$$

**2. Matriks Augmented**

$$
\begin{bmatrix}
1 & 1 & 1 & | & 6 \\
2 & 1 & -1 & | & 3 \\
1 & -1 & 1 & | & 2
\end{bmatrix}
$$



**3. Eliminasi Gauss (Manual)**

**Langkah 1**

Membuat semua elemen di bawah pivot pertama menjadi 0

Operasi:

$
R_2 \leftarrow R_2 - 2R_1
$

$
R_3 \leftarrow R_3 - R_1
$

$$
\begin{bmatrix}
1 & 1 & 1 & | & 6 \\
0 & -1 & -3 & | & -9 \\
0 & -2 & 0 & | & -4
\end{bmatrix}
$$

**Langkah 2**

$
R_3 \leftarrow R_3 + 2R_2
$

$$
\begin{bmatrix}
1 & 1 & 1 & | & 6 \\
0 & -1 & -3 & | & -9 \\
0 & 0 & 6 & | & 14
\end{bmatrix}
$$



**4. Substitusi Balik**

$$
\begin{aligned}
6z &= 14 \Rightarrow z = \frac{7}{3} \\
-y - 3z &= -9 \Rightarrow y = 2 \\
x + y + z &= 6 \Rightarrow x = \frac{5}{3}
\end{aligned}
$$



**5. Solusi**

Jadi, nilai x, y, dan z adalah: 

$$
x = \frac{3}{5}, \quad
y = 2, \quad
z = \frac{3}{7}
$$

**B. Sistem Persamaan Linear Empat Variabel.**

**Soal**

$$
\begin{cases}
x + y + z + w = 10 \\
2x + y - z + w = 8 \\
x - y + 2z - w = 3 \\
3x + 2y + z + w = 14
\end{cases}
$$

**1️⃣ Bentuk Matriks Augmented**

$$
\left[
\begin{array}{cccc|c}
1 & 1 & 1 & 1 & 10 \\
2 & 1 & -1 & 1 & 8 \\
1 & -1 & 2 & -1 & 3 \\
3 & 2 & 1 & 1 & 14
\end{array}
\right]
$$


**2️⃣ Eliminasi Gauss**

**Langkah 1: Hilangkan elemen di bawah pivot pertama (kolom 1)**

$
R_2 \leftarrow R_2 - 2R_1
$

$
R_3 \leftarrow R_3 - R_1
$

$
R_4 \leftarrow R_4 - 3R_1
$

Hasil:

$$
\left[
\begin{array}{cccc|c}
1 & 1 & 1 & 1 & 10 \\
0 & -1 & -3 & -1 & -12 \\
0 & -2 & 1 & -2 & -7 \\
0 & -1 & -2 & -2 & -16
\end{array}
\right]
$$

**Langkah 2: Hilangkan elemen di bawah pivot kedua**

$
R_3 \leftarrow R_3 + 2R_2
$

$
R_4 \leftarrow R_4 + R_2
$

Hasil:

$$
\left[
\begin{array}{cccc|c}
1 & 1 & 1 & 1 & 10 \\
0 & -1 & -3 & -1 & -12 \\
0 & 0 & 7 & 0 & 17 \\
0 & 0 & 1 & -1 & -4
\end{array}
\right]
$$

**Langkah 4: Hilangkan elemen di bawah pivot ketiga**

R4 ​← R4 ​− 7/1​R3​

Hasil akhir bentuk segitiga atas:

$$
\left[
\begin{array}{cccc|c}
1 & 1 & 1 & 1 & 10 \\
0 & 1 & 3 & 1 & 12 \\
0 & 0 & 7 & 0 & 17 \\
0 & 0 & 0 & -1 & -45/7
\end{array}
\right]
$$

**3️⃣ Substitusi Balik**

Dari baris ke-4:

$$
-w = -45/7
$$
$$
w = 45/7
$$



Baris ke-3:

$$
7z = 17
$$
$$
z = \frac{17}{7}
$$



Baris ke-2:

$$
y + 3z + w = 12
$$

Substitusi:

$$
y + 3\left(\frac{17}{7}\right) + \frac{45}{7} = 12
$$

$$
y + \frac{51}{7} + \frac{45}{7} = 12
$$

$$
y + \frac{96}{7} = 12
$$

$$
y = 12 - \frac{96}{7}
$$

$$
y = \frac{84}{7} - \frac{96}{7}
$$

$$
y = -\frac{12}{7}
$$



Baris pertama:

$$
x + y + z + w = 10
$$

Substitusi:

$$
x - \frac{12}{7} + \frac{17}{7} + \frac{45}{7} = 10
$$

$$
x + \frac{50}{7} = 10
$$

$$
x = 10 - \frac{50}{7}
$$

$$
x = \frac{70}{7} - \frac{50}{7}
$$

$$
x = \frac{20}{7}
$$



**=> Jawaban Akhir**

Jadi nilai dari variabel x, y, z, dan w adalah:

$$
x = \frac{20}{7}, \quad
y = -\frac{12}{7}, \quad
z = \frac{17}{7}, \quad
w = \frac{45}{7}
$$
